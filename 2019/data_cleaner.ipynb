```python
import pandas as pd
from hanziconv import HanziConv
```

### This code is to clean data and to merge it from different Wikidata .tsv downloads

```python
lakes=pd.read_csv('lakes_not_preprocessed.tsv',sep='\t')
```

```python
lakes.head()
```

```python
#clean the duplicates
#types = ['lake', 'island', 'mountain', 'river', 'person', 'settlement']
lakes = lakes.drop_duplicates(subset='wiki_id')
```

```python
lakes.shape
```

### Cleaning lakes from 湖 and Russian 'ozero'

```python
lakes = lakes.replace(to_replace=r'(.*)湖$', value=r'\1', regex=True)
```

```python
#no_ozero_t = no_lake_t.replace(to_replace=r'(.*)ское озеро', value=r'\1', regex=True)
#or 
lakes = lakes.replace(to_replace=r'(.*) \(.*\)', value=r'\1', regex=True)
lakes = lakes.replace(to_replace=r'(.*) (озеро|водохранилище)', value=r'\1', regex=True)
lakes = lakes.replace(to_replace=r'Озеро (.*)', value=r'\1', regex=True)
```

```python
lakes[:30]
```

```python
lakes["label_zh"] = lakes["label_zh"].apply(lambda x: HanziConv.toSimplified(x))
lakes[:30]
```

```python
lakes_2=lakes.drop_duplicates(subset=['label_zh', 'label_ru'], keep='last')
lakes_2.shape
```

```python
lakes.to_csv('newlakes.csv')
```

## Settlements

```python
settlements = pd.read_csv('settlements_not_preprocessed.tsv',sep='\t')
settlements.head()
```

```python
settlements.shape
```

```python
settlements = settlements.drop_duplicates(subset='wiki_id')
settlements.shape
```

```python
settlements[:20]
```

```python
settlements = settlements.replace(to_replace=r' \(.*?\)$', value=r'', regex=True)
settlements = settlements.replace(to_replace=r', .*?$', value=r'', regex=True)

settlements[:20]
```

```python
settlements = settlements.replace(to_replace=r' ?(Г|г)о(родской (округ )?)?совет ?', value=r'', regex=True)
settlements = settlements.replace(to_replace=r' ?(С|с)ель(совет|ское поселение) ?', value=r'', regex=True)
settlements = settlements.replace(to_replace=r' ?(М|м)униципальное образование ?', value=r'', regex=True)
settlements = settlements.replace(to_replace=r' ?(Р|р)айон ?', value=r'', regex=True)
settlements = settlements.replace(to_replace=r' ?(У|у)лус ?', value=r'', regex=True)
settlements = settlements.replace(to_replace=r' ?((П|п)о)?с(е|ё)ло($|к) ?', value=r'', regex=True)
settlements = settlements.replace(to_replace=r'(镇)?市(镇)?$', value=r'', regex=True)
settlements = settlements.replace(to_replace=r'(基)?(區|区)$', value=r'', regex=True)
settlements = settlements.replace(to_replace=r'(农)?村(居民点|委员会)?$', value=r'', regex=True)
settlements[:20]
```

```python
settls_2=settlements.drop_duplicates(subset=['label_zh', 'label_ru'], keep='last')
settls_2.shape
```

```python
settls_2["label_zh"] = settls_2["label_zh"].apply(lambda x: HanziConv.toSimplified(x))
settls_2[:20]
```

```python
import re
def to_name_surname(a, b):
    '''
    Converts all personal names to unified order: (First Name)? (Patronym)? Surname
    '''
    if a.count('·')==1:
        if b.count(' ') == 2:
            b = re.sub(r'(.*?) .*? (.*?)', r'\1 \2', b)
    #elif a.count('·')==2:
    #    if b.count(' ') == 1:
    #        a = re.sub(r'(.*?)·.*?·(.*?)', r'\1·\2', a)
    elif a.count('·') == 0:
        if b.count(' ') > 0:
            b = b.split(' ')[-1]
    return b
persons['label_ru'] = persons[['label_zh', 'label_ru']].apply(lambda x: to_name_surname(*x), axis=1)

```
