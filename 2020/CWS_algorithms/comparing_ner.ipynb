# Importing Segmenters

```python
from wordlist_segmenter import wordlist_segmenter #this is our segmenter based on word list from two dictionaries (see README)
with open('long_propers_list.txt', 'r', encoding='utf-8') as f:
    long_propers_list = f.read().split('\n')
    
import jieba #this is a python segmenter from this library: https://github.com/fxsjy/jieba

import pkuseg #this is a new segmenter from Peking University. Can also be found as a python library: https://github.com/lancopku/pkuseg-python
pku_seg = pkuseg.pkuseg()

from nltk.parse.corenlp import CoreNLPParser #this is Stanford NLP parser. To run it, firstly install the CoreNLP software: https://stanfordnlp.github.io/CoreNLP/
stanford_seg = CoreNLPParser('http://localhost:9001')
```

### Segmenting the sentences with all the segmenters

```python
df_news['wordlist'] = df_news['sentence_zho'].apply(lambda x: wordlist_segmenter(x, long_propers_list))
df_books['wordlist'] = df_books['sentence_zho'].apply(lambda x: wordlist_segmenter(x, long_propers_list))
```

```python
df_news['jieba'] = df_news['sentence_zho'].apply(lambda x: '_'.join(list(jieba.cut(x))))
df_books['jieba'] = df_books['sentence_zho'].apply(lambda x: '_'.join(list(jieba.cut(x))))
```

```python
df_news['pku'] = df_news['sentence_zho'].apply(lambda x: '_'.join(pku_seg.cut(x)))
df_books['pku'] = df_books['sentence_zho'].apply(lambda x: '_'.join(pku_seg.cut(x)))
```

```python
df_news['stanford'] = df_news['sentence_zho'].apply(lambda x: '_'.join(list(stanford_seg.tokenize(x))))
df_books['stanford'] = df_books['sentence_zho'].apply(lambda x: '_'.join(list(stanford_seg.tokenize(x))))
```

### Making evaluation functions

The `make_vector` function makes a vector that counts all detected names for a column (i.e. the result of one segmenter).

```python
import re

def make_vector(string):
    n = 0
    
    if string == '':
        n = 0
        
    else:
        res = re.findall(r'_', string)
        n_ = len(res)
        n = n_ + 1
    
    return n
```

Then we make such vectors for each of the columns

```python
df_news['n_real'] = df_news['proper_zh'].apply(lambda x: make_vector(x))
df_news['n_wordlist'] = df_news['wordlist'].apply(lambda x: make_vector(str(x)))
df_news['n_jieba'] = df_news['jieba'].apply(lambda x: make_vector(str(x)))
df_news['n_pku'] = df_news['pku'].apply(lambda x: make_vector(str(x)))
df_news['n_stanford'] = df_news['stanford'].apply(lambda x: make_vector(str(x)))
```

```python
df_books['n_real'] = df_books['proper_zh'].apply(lambda x: make_vector(x))
df_books['n_wordlist'] = df_books['wordlist'].apply(lambda x: make_vector(str(x)))
df_books['n_jieba'] = df_books['jieba'].apply(lambda x: make_vector(str(x)))
df_books['n_pku'] = df_books['pku'].apply(lambda x: make_vector(str(x)))
df_books['n_stanford'] = df_books['stanford'].apply(lambda x: make_vector(str(x)))
```

## Test 1. Cosine similarity

Firstly, we compare the vector of each segmenter with the vector of the true answers.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
news_cos_sim_wordlist = cosine_similarity(df_news['n_real'].to_numpy().reshape(1, -1), df_news['n_wordlist'].to_numpy().reshape(1, -1))[0][0]
news_cos_sim_jieba = cosine_similarity(df_news['n_real'].to_numpy().reshape(1, -1), df_news['n_jieba'].to_numpy().reshape(1, -1))[0][0]
news_cos_sim_pku = cosine_similarity(df_news['n_real'].to_numpy().reshape(1, -1), df_news['n_pku'].to_numpy().reshape(1, -1))[0][0]
news_cos_sim_stanford = cosine_similarity(df_news['n_real'].to_numpy().reshape(1, -1), df_news['n_stanford'].to_numpy().reshape(1, -1))[0][0]
```

```python
books_cos_sim_wordlist = cosine_similarity(df_books['n_real'].to_numpy().reshape(1, -1), df_books['n_wordlist'].to_numpy().reshape(1, -1))[0][0]
books_cos_sim_jieba = cosine_similarity(df_books['n_real'].to_numpy().reshape(1, -1), df_books['n_jieba'].to_numpy().reshape(1, -1))[0][0]
books_cos_sim_pku = cosine_similarity(df_books['n_real'].to_numpy().reshape(1, -1), df_books['n_pku'].to_numpy().reshape(1, -1))[0][0]
books_cos_sim_stanford = cosine_similarity(df_books['n_real'].to_numpy().reshape(1, -1), df_books['n_stanford'].to_numpy().reshape(1, -1))[0][0]
```

## Test 2. Calculating simple ratios

Then we calculate the ratio: (total number of loans detected by segmenter) / (total number of real loans)

```python
news_sum_real = df_news['n_real'].sum()

news_wordlist_ratio = df_news['n_wordlist'].sum() / news_sum_real
news_jieba_ratio = df_news['n_jieba'].sum() / news_sum_real
news_pku_ratio = df_news['n_pku'].sum() / news_sum_real
news_stanford_ratio = df_news['n_stanford'].sum() / news_sum_real
```

```python
books_sum_real = df_books['n_real'].sum()

books_wordlist_ratio = df_books['n_wordlist'].sum() / books_sum_real
books_jieba_ratio = df_books['n_jieba'].sum() / books_sum_real
books_pku_ratio = df_books['n_pku'].sum() / books_sum_real
books_stanford_ratio = df_books['n_stanford'].sum() / books_sum_real
```

## Visualisation in a table

```python
data = {'books: cosine_similarity': [books_cos_sim_wordlist, books_cos_sim_jieba, books_cos_sim_pku, books_cos_sim_stanford],
      'books: overall ratio': [books_wordlist_ratio, books_jieba_ratio, books_pku_ratio, books_stanford_ratio],
      'newspapers: cosine_similarity': [news_cos_sim_wordlist, news_cos_sim_jieba, news_cos_sim_pku, news_cos_sim_stanford],
      'newspapers: overall ratio': [news_wordlist_ratio, news_jieba_ratio, news_pku_ratio, news_stanford_ratio]
     }

df_statistics = pd.DataFrame.from_dict(data, orient='index', columns=['Word List', 'Jieba', 'PKU', 'Stanford']).T
```
